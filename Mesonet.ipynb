```python
def get_weather_features():
    start_date = f"{date.today() - pd.Timedelta(days=9):%Y%m%d}" + '1000'
    end_date = f"{date.today():%Y%m%d}" + "1000"
    
    params = {
    'token': mesowest_auth_key,
    'stid': 'PHOG',
    'start': start_date, # 10:00 to offset UTC time 
    'end': end_date, # 10:00 to offset UTC time
    'obtimezone': 'local',
    'showemptystations': 0,
    'units': 'english',
    'vars': ['wind_speed', 'wind_direction','wind_gust', 'precip_accum_24_hour', 'precip_accum_24_hour_manual',
            'air_temp_high_24_hour', 'air_temp_low_24_hour', 'surface_temp', 'air_temp', 'precip_accum_six_hour',
            'precip_accum_one_hour', 'precip_accum_three_hour']
    }
    weather_data = requests.get('https://api.synopticdata.com/v2/stations/timeseries', params=params).json()
    df = pd.DataFrame(weather_data['STATION'][0]['OBSERVATIONS'])
    df['date_time'] = pd.to_datetime(df['date_time']).dt.date
    df.drop(df[df['date_time'] == date.today()].index, inplace=True)
    
    df_new = pd.DataFrame()

    df_new[['USW00022516_WSF2','USW00022516_WSF5']] = df[['wind_gust_set_1', 'wind_speed_set_1', 'date_time']].groupby('date_time').mean()
    df_new['USW00022516_WDF2'] = df['wind_direction_set_1'].loc[df.groupby('date_time')['wind_gust_set_1'].idxmax().dropna()].to_list()
    df_new['USW00022516_WDF5'] = df['wind_direction_set_1'].loc[df.groupby('date_time')['wind_speed_set_1'].idxmax().dropna()].to_list()
    df_new[['USW00022516_TMIN','USW00022516_TMAX']] = df[['date_time','air_temp_low_24_hour_set_1','air_temp_high_24_hour_set_1']].groupby('date_time').mean()
    df_new[['USW00022516_AWND', 'USW00022516_TAVG']] = df[['date_time', 'wind_speed_set_1', 'air_temp_set_1']].groupby('date_time').mean()
    df_new['USW00022516_PRCP'] = df[['date_time','precip_accum_six_hour_set_1']].groupby('date_time').sum()

    params = {
    'token': mesowest_auth_key,
    'stid': 'COOPMABH1',
    'start': start_date, # 10:00 to offset UTC time 
    'end': end_date, # 10:00 to offset UTC time
    'obtimezone': 'local',
    'showemptystations': 0,
    'units': 'english',
    'vars': ['precip_accum_24_hour', 'precip_accum_six_hour',
            'precip_accum_one_hour', 'precip_accum_three_hour']
    }
    weather_data = requests.get('https://api.synopticdata.com/v2/stations/timeseries', params=params).json()
    df = pd.DataFrame(weather_data['STATION'][0]['OBSERVATIONS'])
    df['date_time'] = pd.to_datetime(df['date_time']).dt.date
    df.drop(df[df['date_time'] == date.today()].index, inplace=True)
    
    df_new['USC00515715_PRCP'] = df[['date_time', 'precip_accum_24_hour_set_1']].groupby('date_time').mean()
    
    params = {
    'token': mesowest_auth_key,
    'stid': 'COOPKBSH1',
    'start': start_date,
    'end': end_date,
    'obtimezone': 'local',
    'showemptystations': 0,
    'units': 'english',
    }
    weather_data = requests.get('https://api.synopticdata.com/v2/stations/timeseries', params=params).json()
    df = pd.DataFrame(weather_data['STATION'][0]['OBSERVATIONS'])
    df['date'] = pd.to_datetime(df['date_time']).dt.date
    df.drop(df[df['date_time'] == date.today()].index, inplace=True)
    
    df_new[['USC00515000_TMIN', 'USC00515000_TMAX', 'USC00515000_PRCP', 'USC00515000_TOBS']] = df[['date', 'air_temp_low_24_hour_set_1', 'air_temp_high_24_hour_set_1', 'precip_accum_24_hour_set_1', 'air_temp_set_1']].groupby('date').mean()
    
    # Fill in NaNs, if they exist
    df_new.interpolate('linear', limit=2, inplace=True)
    df_median = pd.read_csv('weather_features_medians.csv', index_col=0)
    medians = df_median['0'].to_dict()
    df_new.fillna(medians, inplace=True)
    
    return df_new[['USC00515000_PRCP', 'USC00515000_TMAX', 'USC00515000_TMIN', 'USC00515000_TOBS', 
                   'USC00515715_PRCP', 'USW00022516_PRCP', 'USW00022516_TMAX', 'USW00022516_TMIN', 
                   'USW00022516_TAVG', 'USW00022516_WSF2', 'USW00022516_WSF5', 'USW00022516_WDF2', 
                   'USW00022516_AWND', 'USW00022516_WDF5']]
#    return weather_data
```
