```python
stations = ['51000','51001','51002','51003','51004','51101','51213','klih1','kwhh1']
df_all_buoy = pd.DataFrame()
## Needs more testing before deployment. Certain variables are invalid if they are 99 or 999, 
## taking means keeps those data in the dataset when they shouldn't be.


for station in stations:
    path = 'buoy_data/' + station + '/'
    files = os.listdir(path)
#    df_station = pd.DataFrame()
    df_station_agg = pd.DataFrame()
    for file in files:
        df_buoy = pd.read_csv(path + file, delim_whitespace=True, skiprows=[1,2])
        df_buoy.replace([99, 999, 9999], None, inplace=True)
        df_buoy.rename(columns = {'#YY': 'year', 'MM': 'month', 'DD': 'day'}, inplace=True)
        df_buoy['date'] = pd.to_datetime(df_buoy[['year','month','day']])
        df_buoy.set_index('date', inplace=True)
        df_buoy_agg = df_buoy.groupby('date').agg([np.mean, np.std])
        df_buoy_agg.drop([ label1 for label1, label2 in [ col for col, val in df_buoy_agg.sum().items() if val == 0 ] ] \
                         + ['hh', 'mm'], axis=1, inplace=True)
        df_buoy_agg.columns = df_buoy_agg.columns.map('_'.join)
        df_buoy_agg.rename(lambda x: station + '_' + x, axis='columns', inplace=True)
#        df_station = pd.concat([df_station, df_buoy], axis=0)
        df_station_agg = pd.concat([df_station_agg, df_buoy_agg], axis=0)
#    df_station.to_csv(station+'_buoy_data.csv')
    df_station_agg.to_csv(station+'_buoy_data_agg.csv')
    df_all_buoy = pd.concat([df_all_buoy, df_station_agg], axis=1)

df_all_buoy.sort_index(inplace=True)
df_all_buoy.to_csv('buoy_data_agg.csv')
df_all_buoy
```

```python
years = [2017,2018,2019,2020,2021,2022]
months = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']

years_2016 = [2016]
months_2016 = ['06', '07', '08', '09', '10', '11', '12']

years_2023 = [2023]
months_2023 = ['01', '02', '03']

df_rainfall = pd.DataFrame()

for year in years_2016:
    for month in months_2016:
        path = 'rain_data/' + str(year) + '/' + str(month) + '/'
        file = 'rainfall_new_day_statewide_partial_station_data_' + str(year) + '_' + str(month) + '.csv'
        temp = pd.read_csv(path + file)
        temp = temp.query("Island == 'MA'").drop(columns = ['Station.Name', 'Observer', 'Network', \
                                                'Island', 'ELEV.m.','LAT', 'LON', 'NCEI.id', 'NWS.id',\
                                                'NESDIS.id', 'SCAN.id','SMART_NODE_RF.id'])
        temp = temp.set_index('SKN').transpose()
        temp = temp.rename(lambda x: pd.to_datetime(x.replace('X','')))
        df_rainfall = pd.concat([df_rainfall, temp], axis=0)

```

```python
for year in years_2023:
    for month in months_2023:
        path = 'rain_data/' + str(year) + '/' + str(month) + '/'
        file = 'rainfall_new_day_statewide_partial_station_data_' + str(year) + '_' + str(month) + '.csv'
        temp = pd.read_csv(path + file)
        temp = temp.query("Island == 'MA'").drop(columns = ['Station.Name', 'Observer', 'Network', \
                                                'Island', 'ELEV.m.','LAT', 'LON', 'NCEI.id', 'NWS.id',\
                                                'NESDIS.id', 'SCAN.id','SMART_NODE_RF.id'])
        temp = temp.set_index('SKN').transpose()
        temp = temp.rename(lambda x: pd.to_datetime(x.replace('X','')))
        df_rainfall = pd.concat([df_rainfall, temp], axis=0)
```
